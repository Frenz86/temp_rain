```python
import pandas as pd
df = pd.read_excel('tot.xlsx')
df
```

```python
df['Lat'] = df['Lat'].round(6)
df['Lat'] = df['Lat'].astype(str)
df['Lat'] = [x.replace('.', ',') for x in df['Lat']]
#df['Lat'] = df['Lat'].astype(float)
df
```

```python
df['Long'] = df['Long'].round(6)
df['Long'] = df['Long'].astype(str)
df['Long'] = [x.replace('.', ',') for x in df['Lat']]
#df['Lat'] = df['Lat'].astype(float)
df
```

```python
df.to_excel('tot_converted.xlsx')
```
